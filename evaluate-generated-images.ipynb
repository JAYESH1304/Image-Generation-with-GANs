**Structural Similarity Index Measure**

```python
def max_similar_image(image_path):
    
    image1 = imageio.imread(image_path)
    base_path = '/kaggle/input/healthy/Healthy/'
    
    SSIM = 0
    index = 0
    
    for i in range(1,365):
        path = os.path.join(base_path,f'image_{i:03d}.png')
        image2 = imageio.imread(path)
        image2 = cv2.resize(image2, (64, 64))

        ssim_generated = ssim(image1, image2, win_size=5, channel_axis=-1)
        
        if ssim_generated > SSIM:
            SSIM = ssim_generated
            index = i
            
    return SSIM
#     path = os.path.join(base_path,f'image_{index:03d}.png')
#     image2 = imageio.imread(path)
#     image2 = cv2.resize(image2, (64, 64))
    
#     fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 4),
#                          sharex=True, sharey=True)
#     ax = axes.ravel()
    
#     ax[0].imshow(image2, vmin=0, vmax=1)
#     ax[0].set_xlabel('SSIM: 1.00')
#     ax[0].set_title('Original Image')

#     ax[1].imshow(image1, vmin=0, vmax=1)
#     ax[1].set_xlabel(f'SSIM: {SSIM:.2f}')
#     ax[1].set_title('Generated Image')

#     plt.tight_layout()
#     plt.show()
```

**Peak Signal-to-Noise Ratio**

```python
import imageio
import os
import numpy as np
import math
import cv2

def psnr(img1, img2):
    mse = np.mean((img1 - img2) ** 2)
    if mse == 0:
        return float('inf')  # Infinite PSNR means identical images
    max_pixel = 255.0
    psnr_value = 20 * math.log10(max_pixel / math.sqrt(mse))
    return psnr_value

def max_similar_image_psnr(image_path):
    base_path = '/kaggle/input/healthy/Healthy/'
    target_image = imageio.imread(image_path)

    max_psnr = 0
    max_index = 0

    for i in range(1, 366):
        path = os.path.join(base_path, f'image_{i:03d}.png')
        comparison_image = imageio.imread(path)
        comparison_image = cv2.resize(comparison_image, (64, 64))

        current_psnr = psnr(target_image, comparison_image)
        
        if current_psnr > max_psnr:
            max_psnr = current_psnr
            max_index = i

    return max_psnr, max_index
```

```python
psnr_value, image_index = max_similar_image_psnr("/kaggle/input/generated-healthy/Generated_Images/kaggle/working/final_365/sharped_image_00027.png")
print(f"Max PSNR: {psnr_value}, Image Index: {image_index}")
```
